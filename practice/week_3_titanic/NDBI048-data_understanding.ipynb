```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = [8, 6]
sns.set_theme(style="ticks", color_codes=True)
```

```python
df: pd.DataFrame = pd.read_csv('titanic_train.csv')
data: pd.DataFrame = df.groupby('ticket').agg(pclass=('pclass', 'max'), fare=('fare', 'max'), age=('age', 'max'))
```

# Histograms in python

```python
fig, ax = plt.subplots(2, 3, figsize=(14, 10))

# continuous variable
sns.histplot(data=df, x='age', ax=ax[0, 0])
sns.histplot(data=df, x='age', bins=10, ax=ax[0, 1])
sns.histplot(data=df, x='age', binwidth=10, ax=ax[0, 2])

sns.histplot(data=df, x='age', binrange=(0, 25), ax=ax[1, 0])
sns.histplot(data=df, x='age', log_scale=(False, 2), ax=ax[1, 1])
sns.histplot(data=df, x='age', stat='probability', ax=ax[1, 2]);
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 6))

# discrete variable
sns.histplot(data=df, x='pclass', discrete=True, shrink=.9, ax=ax[0])
sns.countplot(data=df, x='pclass', ax=ax[1]);
```
